-> **generators** are type function that return  an iterator.

-> It uses yield instead of return.


```python
def count_up(num):
   i=0
   while i<num:
    yield i
    i+=1
for j in count_up(6):
    print(j)
```

# 7- What are the advantages of using generators over regular functions?

->Memory efficient

->Lazy evaluation

->Faster for large data

->No need to store all values

```python
def even(n):
    for i in range(n):
        yield i  # yield pauses the execution of the generator after producing a value
        
h=even(11)
# every next() call we resumes the generator for next execution
print(next(h))
print(next(h))
print(next(h))
```

# Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
nums = [10, 15, 20, 25, 30]
def sum_of_even(numbers):
    total=0
    for i in numbers:
        if i%2==0:
            total+=i
    return total

result = sum_of_even(nums)
print(result)
```

# Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n <= 1:
        return False
    return len(list(filter(lambda x: n % x == 0, range(2, n)))) == 0


primes_1_to_200 = list(filter(is_prime, range(1, 201)))
print(primes_1_to_200)
```

# Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms      
        self.count = 0          
        self.a = 0              
        self.b = 1              
    def __iter__(self):
        return self             

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration

        value = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return value

fib = Fibonacci(7)

for num in fib:
    print(num)
```

# Write a generator function in Python that yields the powers of 2 up to a given exponent

```python
def power_of_two(n):
    for i in range(n+1):
        yield 2**i

num=int(input("Enter the number :"))

for j in power_of_two(num):
    print(j)
```

#  Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(filename):
    with open(filename,mode="r") as f:
        for line in f:
            yield line.strip()
```

#  Create a Python program that uses `filter()` to remove all the vowels from a given string

```python
def remove_vowels(text):
    vowels="aeiou"
    return "".join(list(filter(lambda x: x.lower() not in vowels,text)))

remove_vowels("Hi")
```

# Write a Python program using lambda and map.

```python
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]

result = list(
    map(
        lambda order: (
            order[0],
            order[2] * order[3] + 10 if order[2] * order[3] < 100 else order[2] * order[3]
        ),
        orders
    )
)

print(result)
```
