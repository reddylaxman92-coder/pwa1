**Tuples**  are stored in memory with fixed blocks. Python locks it's structure for safety and efficieny. So we can not update ,remove, add any new element inside tuples directly. If we are using list inside tuple then we can update that list separately.

### **12- How does the elif statement work ?**

The elif (else if) statement is used when you want to check multiple conditions one by one.

Python starts by checking the if condition. If the if condition is False, it moves to the elif condition.
If the elif condition is True, it executes that block.
If both if and all elif conditions are False, Python finally executes the else block.

# Write a function to demonstrate the behavior of mutable and immutable arguments

```python
def update_value(x):       ## it will not affect to original varriable because when we assign any value it creats new memory object. So it is immutable
  print("Inside function (immutable):", id(x))
  x=x+10
  print("Inside function (immutable):", id(x))


def update_list(lst):     ## it will  affect to original list because it allocates memory with extra spaces by which we update ,add and remove any item. Which makes the list as mutable.
    print("Before change (mutable):", id(lst))
    lst.append(100)   # modifies SAME list in place
    print("After change  (mutable):", id(lst))

update_value(20)

print("---------------------------------------------")
my_list = [10, 20, 30]
update_list(my_list)
```

 #  Write code to demonstrate type casting with list elements.

```python
data = ["10", "20", "30", "40", "50"]

integer_value=[]
boolean_value=[]
float_value =[]

for d in data:
  integer_value.append(int(d))
  boolean_value.append(bool(d))
  float_value.append(float(d))

print(integer_value)
print(boolean_value)
print(float_value)

```

#  Write a program to reverse a string using a while loop.

```python
myStr = "Data Science"
reversed_string=""

i= len(myStr)-1

while i>=0:
  reversed_string+=myStr[i]
  i-=1
print(reversed_string)
```

# Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
number = int(input("Enter number:"))
i=1
factorial=1

while i<=number:
  factorial*=i
  i+=1
print(factorial)
```
